```python
import requests
from bs4 import BeautifulSoup
```

```python
main = requests.get("https://www.congress.gov/sponsors-cosponsors/116th-congress/senators/all")
```

```python
soup = BeautifulSoup(main.content, 'html.parser')
#print(soup.prettify())
main_table = soup.find(class_="table-wrapper")
print(main_table.prettify())
```

```python
senator_names=[]
senator_links=[]
senator_count = 1
for senator in main_table.find_all('tr'):
    #print(senator.prettify())
    for name in senator.find_all('td', {"data-text": True}):
        #print(name.get_text())
        senator_names.append(name.get_text())
        #senators_dict.update({"Senator "+str(senator_count):})
        #print("https://www.congress.gov"+name.find("a")['href'])
        senator_links.append("https://www.congress.gov"+name.find("a")['href'])
     #   senator_dict = {}
     #   senator_names.append(name.get_text())
     #   senator_dict.update({'name':name.get_text()})
     #   senator_names_links.append("https://www.congress.gov"+ name.find("a")['href'])
     #   senator_dict.update({'link':"https://www.congress.gov"+ name.find("a")['href']})
     #   senators_dicts.update({name.get_text():senator_dict})
    senator_count += 1

print(senator_names)
print(senator_links)
print(len(senator_names))
```

```python
senators_dictionaries_list = []
cuenta_for = 0
for link in senator_links:

    senator_bills_dict = {}

    sponsored = requests.get(link+'?q=%7B"sponsorship"%3A"sponsored"%2C"bill-status"%3A"law"%7D')
    senator_soup = BeautifulSoup(sponsored.content, 'html.parser')
    #print(senator_soup.prettify())
    
    #senator_bill_table = senator_soup.find_all('ol', class_="basic-search-results-lists expanded-view")
    
    try:
        senator_bill_table = senator_soup.find_all('ol', class_="basic-search-results-lists expanded-view")
        prueba = senator_bill_table[0].find_all('li', class_="expanded")
    except:
        print("The senator number "+str(cuenta_for)+ " has passed no bills")
        
        senator_bills_dict.update({"Bill":"N/A"})
        senators_dictionaries_list.append(senator_bills_dict)
        cuenta_for += 1
        continue
        
    #print(senator_bill_table[0].get_text())
    bill_count = 1
    for bill in senator_bill_table[0].find_all('li', class_="expanded"):
        #print(bill.get_text())
        bill_dict = {}
        bill_number = bill.find('span', class_="result-heading").find('a')
        #print(bill_number.get_text())
        bill_dict.update({"Bill Number":bill_number.get_text()})

        bill_cosponsors = bill.find('span', class_="result-item").find_all('a')[1]

        bill_dict.update({"Number of Cosponsors":bill_cosponsors.get_text()})

        cosponsors_dict = {}

        #print("Cosponsors: " + bill_cosponsors.get_text())
        #print('https://www.congress.gov'+bill_cosponsors['href'])

        if int(bill_cosponsors.get_text()) > 0:

            #print("Cosponsors:")

            cosponsors_link = requests.get('https://www.congress.gov'+bill_cosponsors['href'])
            cosponsors_soup = BeautifulSoup(cosponsors_link.content, 'html.parser')
            print('https://www.congress.gov'+bill_cosponsors['href'])
            name_table = cosponsors_soup.find('table', class_='item_table').find('tbody')
            #print(name_table.prettify())

            #print(int(bill_cosponsors.get_text()))
            count = 1
            for cosponsor in name_table.find_all('tr'):
                cosponsor_info = cosponsor.find('td').find('a')
                cosponsor_name = cosponsor_info.get_text()
                cosponsor_link = cosponsor_info['href']
                #print(cosponsor_name + "@" + cosponsor_link)
                cosponsors_dict.update({"Cosponsor "+str(count):cosponsor_name + "@" + cosponsor_link})
                count += 1

        else:
            #print("No Cosponsors")
            cosponsors_dict.update({"Cosponsor":"\\"})


        bill_dict.update({"Cosponsor Info":cosponsors_dict})
        senator_bills_dict.update({"Bill "+str(bill_count):bill_dict})
        bill_count+=1

    #print(senator_bills_dict)
    cuenta_for +=1
    print("Senador terminado "+str(cuenta_for))
    senators_dictionaries_list.append(senator_bills_dict)
    
#senators_dictionaries_list.append(senator_bills_dict)
```
