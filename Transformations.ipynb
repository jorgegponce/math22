```python
def page_rank(t, p, iterations):
    
    dim = len(t)
    s = np.full((dim,dim), 1/dim)
    
    m = (1-p)*t + p*s
    
    v = np.random.rand(len(m), 1)
    v = v / np.linalg.norm(v, 1)
    
    for x in range(iterations):
        v = m @ v
        
    return v
        
```

```python
importance_vector = page_rank(t, 0.15, 1000)
importance_vector
```

```python
print(np.amax(importance_vector))
index = np.argmax(importance_vector)
print(index)
print(importance_vector[index])
importance_vector = importance_vector/np.amax(importance_vector)
print(importance_vector)
```

```python
print(importance_vector)
imp_reshaped = importance_vector.transpose()
print(imp_reshaped[0])
r = np.argsort(imp_reshaped)
print(r[0])
```
